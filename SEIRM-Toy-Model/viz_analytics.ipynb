```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def RMSE(predictions,targets): 
    return np.sqrt(np.mean((predictions-targets)**2))
```

```python
df = pd.read_csv('log.csv')
```

```python
df.tail(3)
```

```python
data = df.loc[(df.ID == 'Noise in D_reported') & (df.noise_level == 0.1)]

fig, axs = plt.subplots(2,1,figsize=(6,6))
for i in range(2):
    axs[i].grid('on')
axs[0].plot(data.Min - data.GT)
axs[0].legend(['Difference GT with Min'])

axs[1].plot(data.Start- data.GT)
axs[1].legend(['Diff GT with Start'])

plt.savefig('SEIRMmodel_diff.png',dpi=300, bbox_inches='tight', pad_inches=0)

plt.show()
```

```python
plt.plot(data.GT,marker='*',label='D_gt')
plt.plot(data.Start,label='D_start')
plt.plot(data.Min,label='D_min')
plt.savefig('SEIRMmodel.png',dpi=300, bbox_inches='tight', pad_inches=0)
plt.legend()
plt.show()
```

```python
RMSE(data.Start , data.GT), RMSE(data.Min,data.GT)
```
